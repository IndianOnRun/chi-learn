```python
import json
import re
```

```python
geoJson = {}
with open('../json/communityAreasGeo.json') as input:
    geoJson = json.load(input)
```

```python
geoJson['features'][0]['properties']
```

```python
pattern = re.compile('<span class=\\"atr-name\\">AREA_NUMBE</span>:</strong> <span class=\\"atr-value\\">(\d+)</span>')
```

```python
pattern
```

```python
for i, feat in enumerate(geoJson['features']):
    area_num = re.search(pattern, feat['properties']['description']).group(1)
    del geoJson['features'][i]['properties']
    geoJson['features'][i]['area_number'] = area_num
    
```

```python
geoJson
```

```python
with open('commAreasGeo.json', 'w') as output:
    json.dump(geoJson, output)
```
