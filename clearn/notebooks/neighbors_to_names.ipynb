```python
import csv
```

```python
with open('../config/community_areas.csv', 'r') as areas:
    area_to_name = {}
    reader = csv.reader(areas)
    for line in reader:
        area_to_name[line[0]] = line[1]
        
area_to_name['1']
```

```python
with open('../config/community_area_neighbors.csv', 'r') as neighbors:
    neighbor_map = {}
    neighbor_reader = csv.reader(neighbors)
    for line in neighbor_reader:
        neighbor_map[line[0]] = line[1:]
        
neighbor_map['1']
```

```python
new_neighbor_map = {}
for area in neighbor_map:
    neighbors = neighbor_map[area]
    new_neighbors_for_area = []
    for n in neighbors:
        new_neighbors_for_area.append(area_to_name[n])
        
    new_neighbor_map[area_to_name[area]] = new_neighbors_for_area
    
new_neighbor_map
```

```python
writer = csv.writer(open('../config/neighbors_as_names.csv', 'w'))
```

```python
for area in new_neighbor_map:
    array_to_write = new_neighbor_map[area]
    array_to_write.insert(0, area)
    writer.writerow(array_to_write)
```
